```python
import pandas as pd

df = pd.read_csv('germany_city_data.csv')

df.head()
```

```python
df['geom'] = 'SRID=4326;POINT(' + df['longitude'].astype(str) + ' ' + df['latitude'].astype(str) + ')'
df.head()
```

```python
df = df.drop(columns=['population_rank', 'longitude', 'latitude'])

df.head()
```
